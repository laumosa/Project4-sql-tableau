### 1. Import libraries

```python
import pymysql
import sqlalchemy as alch
from getpass import getpass
import pandas as pd
```

### 2. Connect to SQL

```python
password = getpass("Please input your password: ")
dbName = "movies_dataset"
connectionData=f"mysql+pymysql://root:{password}@localhost/{dbName}"
engine = alch.create_engine(connectionData)
```

### 3. SQL Queries

```python
query = """SELECT * FROM movies_table
            LIMIT 5;"""
pd.read_sql_query(query, engine)
```

```python
# 1. What are the top 5 highest-rated films of the 21st century?
query = """SELECT decade, name, average_rating FROM movies_table 
           WHERE average_rating IS NOT NULL
           AND decade = '2000s'
           OR decade = '2010s'
           OR decade = '2020s'
           ORDER BY average_rating DESC
           LIMIT 5;"""
pd.read_sql_query(query, engine)
```

```python
# 2. Which film has the longest duration?
query = """SELECT name, duration_min FROM movies_table 
           ORDER BY duration_min DESC
           LIMIT 1;"""
pd.read_sql_query(query, engine)
```

```python
# 3. Who are the directors with more than 9 films in the database?
query = """SELECT director, COUNT(name) FROM movies_table 
           WHERE director IS NOT NULL
           GROUP BY director
           HAVING COUNT(name) > 9
           ORDER BY COUNT(name) DESC;"""
pd.read_sql_query(query, engine)
```

```python
# 4. What is the average rating of films in Horror genre?
query = """SELECT AVG(average_rating) FROM movies_table 
           WHERE average_rating IS NOT NULL
           AND genre LIKE '%%Horror%%';"""
pd.read_sql_query(query, engine)
```

```python
# 5. What is the average rating of films in Comedy genre?
query = """SELECT AVG(average_rating) FROM movies_table 
           WHERE average_rating IS NOT NULL
           AND genre LIKE '%%Comedy%%';"""
pd.read_sql_query(query, engine)
```

```python
# 6. In how many films did Robin Williams appear?
query = """SELECT COUNT(name) FROM movies_table 
           WHERE cast LIKE '%%Robin Williams%%';"""
pd.read_sql_query(query, engine)
```

```python
# 7. How many films were released in each decade?
query = """SELECT decade, COUNT(name) FROM movies_table 
           WHERE decade IS NOT NULL
           GROUP BY decade
           ORDER BY decade;"""
pd.read_sql_query(query, engine)
```

```python
# 8. Which films have an average rating below or equal to 1?
query = """SELECT name, average_rating FROM movies_table 
           WHERE average_rating <= 1;"""
pd.read_sql_query(query, engine)
```
